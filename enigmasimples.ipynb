```python
alphabet = [
    'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
    'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'
]

accent_map = {
    'á': 'a', 'à': 'a', 'ä': 'a', 'â': 'a', 'ã': 'a',
    'é': 'e', 'è': 'e', 'ë': 'e', 'ê': 'e',
    'í': 'i', 'ì': 'i', 'ï': 'i', 'î': 'i',
    'ó': 'o', 'ò': 'o', 'ö': 'o', 'ô': 'o', 'õ': 'o',
    'ú': 'u', 'ù': 'u', 'ü': 'u', 'û': 'u',
    'ñ': 'n', 'ç': 'c',
} # 224 - 230 (a), 231 (ç), 232 - 235 (e), 236 - 239 (i), 241 (ñ), 242 - 246 (o), 249 - 252 (u)

accent_map_ord = {
    224: 'a', 225: 'a', 226: 'a', 227: 'a', 228: 'a', 229: 'a', 230: 'a',
    231: 'c',
    232: 'e', 233: 'e', 234: 'e', 235: 'e',
    236: 'i', 237: 'i', 238: 'i', 239: 'i',
    241: 'n',
    242: 'o', 243: 'o', 244: 'o', 245: 'o', 246: 'o',
    249: 'u', 250: 'u', 251: 'u', 252: 'u'
}

# import time

rangeA = 224
rangeB = 252

def normalize_case(text: str) -> list:
  no_cases = text.lower()
  return no_cases

def normalize_spaces(text: str) -> list:
  no_spaces = text.replace(" ", "")
  return no_spaces

def normalize_by_alphabet(text: str) -> list:
  odd_chars = list()
  for i in text:
    if not i in alphabet:
      odd_chars.append(i)
  return odd_chars

def in_between_v(v: int, a: int, b: int) -> bool:
  if a <= v <= b:
    return True
  return False

def normalize_accents_range(text: str) -> list: # A bit slower. Optimize?
  cleaned = ""
  for c in text:
    if rangeA <= ord(c) <= rangeB:
      cleaned += accent_map_ord[ord(c)]
    else:
      cleaned += c
  return cleaned

def normalize_accents_in(text: str) -> list: # Use this
  cleaned = ""
  for c in text:
    if c in accent_map:
      cleaned += accent_map[c]
    else:
      cleaned += c
  return cleaned

def rem_spec_char(text: str):
  has_special = [char for char in text if char.isalnum()]
  return has_special

def rem_nums(text: str):
  has_numbers = [num for num in text if not num.isdigit()]
  return has_numbers

def clean_text(text: str) -> list:
  no_cases = normalize_case(text)
  no_spaces = normalize_spaces(no_cases)
  no_accents = normalize_accents_in(no_spaces)
  no_spec = rem_spec_char(no_accents)
  no_all = rem_nums(no_spec)
  return "".join(no_all)

#cleaned = clean_text("Fundamentos de Criptografia: é uma disciplina do curso de Sistemas de Informação & Computação")
print(clean_text("Olá mundo"))
#print(cleaned)

```

```python
# Vamos assumir uma configuração padrão para a máquina simplificada do Enigma
# O plugboard não será implementado

''' Instruções:
Implemente qualquer versão da Máquina Enigma.
Não precisa se preocupar com a parte visual.
Para simplificar NÃO PRECISA IMPLEMENTAR o "plugboard" que faz as ligações entre letras do teclado antes de enviar para os rotores.

ENTRADA: O programa recebe uma lista com três números que são os três rotores escolhidos (entre um total de 5), recebe outra lista com três números de 1 a 26 que será a posição inicial de cada rotor. O programa recebe também um txt que é a mensagem (pode assumir que o usuário vai limpar o texto - senão eu reprovo esse usuário)
'''

# Os 5 rotores possíveis, já com a fiação embaralhada de cada um
# Garante que os rotores possam ser reusados pra descriptografia
'''
Configurações usadas:
rotor 1 = EKMFLGDQVZNTOWYHXUSPAIBRCJ; ranhura = Q
rotor 2 = AJDKSIRUXBLHWTMCQGZNPYFVOE; ranhura = E
rotor 3 = BDFHJLCPRTXVZNYEIWGAKMUSQO; ranhura = V
rotor 4 = ESOVPZJAYQUIRHXLNFTGKDCMWB; ranhura = J
rotor 5 = VZBRGITYUPSDNHLXAWMJQOFECK; ranhura = Z
refletor_b = YRUHQSLDPXNGOKMIEBFZCWVJAT
'''
def pega_rotores():
  return [
      "EKMFLGDQVZNTOWYHXUSPAIBRCJ",  # Rotor 1
      "AJDKSIRUXBLHWTMCQGZNPYFVOE",   # Rotor 2
      "BDFHJLCPRTXVZNYEIWGAKMUSQO",   # Rotor 3
      "ESOVPZJAYQUIRHXLNFTGKDCMWB",    # Rotor 4
      "VZBRGITYUPSDNHLXAWMJQOFECK"    # Rotor 5
  ]

REFLETOR_B ="YRUHQSLDPXNGOKMIEBFZCWVJAT"

# Possível configuração pra testes: 3,1,4
rotores_escolhidos = [3,1,4]
'''
ROTOR_1 = ROTORES_CONFIG_FIACAO[rotores_escolhidos[0]]
ROTOR_2 = ROTORES_CONFIG_FIACAO[rotores_escolhidos[1]]
ROTOR_3 = ROTORES_CONFIG_FIACAO[rotores_escolhidos[2]]
'''
# Posições de Avanço (onde fica localizado a ranhura que permite os rotores após o mais a direita rodarem)
# Pra simplificar manteremos essa distribuição
ROTORES_RANHURAS = [ord('Q'), ord('E'), ord('V'), ord('J'), ord('Z')]

# Definir um alfabeto, usando maiúsculas pra alinhar
AZ_alphabet = [l.lower() for l in alphabet]
N_alphabet = {}
C_alphabet = {}

# Mapear as letras numericamente
for index, letter in enumerate(AZ_alphabet):
  N_alphabet.update({index: letter})
  C_alphabet.update({letter: index})

def char_para_int(char):
    # Assumimos que o char já está limpo e minúsculo
    return C_alphabet.get(char)

def int_para_char(int_val):
    return N_alphabet.get(int_val)

def setup_enigma(rotores, pos_iniciais_referencia):
    # CONVERSÃO 1–26 → 0–25
    posicoes_usadas = [(p - 1) % 26 for p in pos_iniciais_referencia]

    ROTORES_CONFIG_FIACAO = pega_rotores()
    if len(rotores) != 3 or len(posicoes_usadas) != 3:
        raise ValueError("São necessários 3 rotores e 3 posições iniciais.")

    rotores_usados = []
    ranhuras = []
    for n in rotores:
        id_rotor = n - 1
        rotores_usados.append(ROTORES_CONFIG_FIACAO[id_rotor].lower())
        ranhuras.append(ROTORES_RANHURAS[id_rotor])

    print(f"--- Configuração Inicial ---")
    print(f"Ordem dos Rotores (E/M/D): {rotores}")
    print(f"Posições Iniciais (internas): {posicoes_usadas}")
    print("----------------------------")

    return rotores_usados, ranhuras, posicoes_usadas

def avancar_rotores(ranhuras, posicoes):
    avanca_esquerda = False
    avanca_meio = False

    # Double-step: o rotor do meio na ranhura
    if posicoes[1] == ranhuras[1]:
        avanca_meio = True
        avanca_esquerda = True

    # O rotor da direita na ranhura empurra o do meio
    elif posicoes[2] == ranhuras[2]:
        avanca_meio = True

    # Rotor da direita sempre avança
    posicoes[2] = (posicoes[2] + 1) % 26

    # Avanços condicionais
    if avanca_meio:
        posicoes[1] = (posicoes[1] + 1) % 26

    if avanca_esquerda:
        posicoes[0] = (posicoes[0] + 1) % 26

    return posicoes

def mapear_rotor(entrada_int, rotor_index, direcao, rotores_config, posicoes):
    # A posição (offset) do rotor atual
    pos = posicoes[rotor_index]
    # A fiação (string) do rotor atual
    rotor_fios = rotores_config[rotor_index].lower() # garantir minúsculas

    # 1. Ajuste de Entrada (Entrada no Fio)
    entrada_ajustada = (entrada_int + pos) % 26

    # 2. Mapeamento
    if direcao == "IDA":
        # Mapeamento Direto:
        # Pega a letra na posição 'entrada_ajustada' da string de fiação
        saida_mapeada = char_para_int(rotor_fios[entrada_ajustada])
    else: # direcao == "VOLTA" (Mapeamento Inverso)
        # Mapeamento Inverso:
        # Procura onde o valor da 'entrada_ajustada' está no rotor_fios
        letra_procurada = int_para_char(entrada_ajustada)
        saida_mapeada = rotor_fios.find(letra_procurada)

    # 3. Ajuste de Saída (Retirada da Posição)
    # Se o erro persiste, é provável que char_para_int/rotor_fios.find() retornou None ou -1.
    # No seu caso, o find retorna -1, que é um int, mas o char_para_int retorna None.

    if saida_mapeada is None: # Se a conversão falhou (maior chance na IDA)
        raise ValueError(f"Erro de Mapeamento: Fiação '{rotor_fios}' não contém a letra procurada.")

    saida_int = (saida_mapeada - pos) % 26
    saida_int = (saida_int + 26) % 26

    return saida_int

def criptografar_letra(letra, rotores_config, ranhuras_config, posicoes_config):
    # 1. A posição muda ANTES da criptografia.
    avancar_rotores(ranhuras_config, posicoes_config)

    sinal_int = char_para_int(letra)

    # 2. PASSAGEM DE IDA (Direita -> Esquerda)
    # A ordem dos rotores no array é [0, 1, 2] -> [R3, R2, R1]
    # R1 (Direita, índice 2)
    sinal_int = mapear_rotor(sinal_int, 2, "IDA", rotores_config, posicoes_config)
    # R2 (Meio, índice 1)
    sinal_int = mapear_rotor(sinal_int, 1, "IDA", rotores_config, posicoes_config)
    # R3 (Esquerda, índice 0)
    sinal_int = mapear_rotor(sinal_int, 0, "IDA", rotores_config, posicoes_config)

    # 3. REFLETOR
    # O refletor tem sempre fiação fixa
    refletor_char = REFLETOR_B[sinal_int].lower() # Garante minúsculas
    sinal_int = char_para_int(refletor_char)

    # 4. PASSAGEM DE VOLTA (Esquerda -> Direita)
    # O sinal retorna no R3 (índice 0)
    # R3 (Esquerda, índice 0)
    sinal_int = mapear_rotor(sinal_int, 0, "VOLTA", rotores_config, posicoes_config)
    # R2 (Meio, índice 1)
    sinal_int = mapear_rotor(sinal_int, 1, "VOLTA", rotores_config, posicoes_config)
    # R1 (Direita, índice 2)
    sinal_int = mapear_rotor(sinal_int, 2, "VOLTA", rotores_config, posicoes_config)

    # 5. Saída Final
    return int_para_char(sinal_int)

def criptografar_descriptografar_texto(texto, escolha_rotores, posicoes_iniciais_de_entrada):
    # A. Configuração e Estado Inicial
    try:
        ROTORES_CONFIG, RANHURAS_CONFIG, POSICOES_CONFIG = setup_enigma(escolha_rotores, posicoes_iniciais_de_entrada)
    except ValueError as e:
        return f"ERRO DE CONFIGURAÇÃO: {e}"

    # B. Limpeza do Texto (Usando sua função)
    texto_limpo = texto
    print(f"Texto Original (Limpo): {texto_limpo}")

    resultado_cifrado = ""

    # C. Criptografia Letra por Letra
    for letra in texto_limpo:
        letra_cifrada = criptografar_letra(
            letra,
            ROTORES_CONFIG,
            RANHURAS_CONFIG,
            POSICOES_CONFIG
        )
        resultado_cifrado += letra_cifrada

    return resultado_cifrado

def iniciar():
  # Escolher os rotores que serão usados:
  ROTORES_DISPONIVEIS = pega_rotores()
  ROTORES_NUMEROS = [rotor+1 for rotor, config in enumerate(ROTORES_DISPONIVEIS)]
  rotor1 = int(input(f"Selecione o primeiro rotor dos disponíveis {ROTORES_NUMEROS}: "))
  ROTORES_NUMEROS.remove(rotor1)
  rotor2 = int(input(f"Selecione o segundo rotor dos disponíveis {ROTORES_NUMEROS}: "))
  ROTORES_NUMEROS.remove(rotor2)
  rotor3 = int(input(f"Selecione o terceiro rotor dos disponíveis {ROTORES_NUMEROS}: "))
  ROTORES_NUMEROS.remove(rotor3)
  rotores_escolhidos = [rotor1, rotor2, rotor3]
  print(f"Os rotores escolhidos foram {rotores_escolhidos}")
  pos1 = int(input(f"Selecione a posição inicial do rotor {rotor1}, entre 1 e 26: "))
  pos2 = int(input(f"Selecione a posição inicial do rotor {rotor2}, entre 1 e 26: "))
  pos3 = int(input(f"Selecione a posição inicial do rotor {rotor3}, entre 1 e 26: "))
  posicoes_iniciais = [pos1, pos2, pos3]
  texto_entrada = str(input("Digite o texto que deseja criptografar: "))
  texto_limpo = clean_text(texto_entrada)
  cip = criptografar_descriptografar_texto(texto_limpo, rotores_escolhidos, posicoes_iniciais)
  print(f"O texto criptografado é: {cip}")
  posicoes_iniciais = [pos1, pos2, pos3]
  dec = criptografar_descriptografar_texto(cip, rotores_escolhidos, posicoes_iniciais)
  print(f"O texto descriptografado é: {dec}")

iniciar()
```
